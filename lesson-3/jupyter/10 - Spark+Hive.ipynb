```python
from pyspark.sql import HiveContext
```

```python
sqlContext = HiveContext(sc)
```

```python
sqlContext.sql("CREATE TABLE IF NOT EXISTS miguxos(cod INT, name STRING, age INT, qtd INT) row format delimited fields terminated by ','")
```

```python
sqlContext.sql("show tables").collect()
```

```python
sqlContext.sql("show databases").collect()
```
